```python
import georasters as gr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

DIF is diffuse horizontal Irradiation measured in kWh/m2, covering the period 1994-2018

```python
DIF = gr.from_file("/DIF.tif")
```

```python
DIF.plot()
```

```python
DIF = DIF.to_pandas()
```

```python
DIF.head()
```

```python
DIF = DIF.loc[:,['value', 'x', 'y']]
DIF.head()
```

```python
DIF = DIF.rename(columns={'value':'DIF','x':'Latitude', 'y':'Longitude'})
DIF.head()
```

```python
DIF.info
```

Longterm yearly average of direct normal irradiation (DNI) in kWh/m2, covering
the period 1994-2018

```python
DNI = gr.from_file("/DNI.tif")
```

```python
DNI.plot()
```

```python
DNI = DNI.to_pandas()
DNI.head()
```

```python
DNI = DNI.loc[:,['value', 'x', 'y']]
DNI = DNI.rename(columns={'value':'DNI','x':'Latitude', 'y':'Longitude'})
DNI.head()
```

Longterm yearly average of global horizontal irradiation (GHI) in kWh/m2,
covering the period 1994-2018

```python
#Do the same process for the the remaining data
GHI = gr.from_file('/GHI.tif')
```

```python
GHI.plot()
```

```python
GHI = GHI.to_pandas()
GHI.head()
```

```python
GHI = GHI.loc[:,['value','x', 'y']]
GHI = GHI.rename(columns={'value':'GHI', 'x':'Latitude','y':'Longitude'})
GHI.head()
```

Longterm yearly average of global irradiation at optimum tilt (GTI) in kWh/m2,
covering the period 1994-2018


```python
GTI = gr.from_file("/GTI.tif")
GTI.plot()
```

```python
GTI = GTI.to_pandas()
GTI = GTI.loc[:,['value','x', 'y']]
GTI = GTI.rename(columns={'value':'GTI', 'x':'Latitude','y':'Longitude'})
GTI.head()
```

```python
dfs = [DIF,DNI,GHI,GTI,OPTA,TEMP]
nan_value = 0
result_1 = pd.concat(dfs, join='outer', axis=1).fillna(nan_value)

```

```python
result_1.head()
```

```python
NIG_SOLAR_IRRADIATION_dataset = result_1.to_csv('NIG_SOLAR_IRRADIATION_dataset.csv')
```
